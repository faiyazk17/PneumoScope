```python
# eda.ipynb
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Set the paths to the data
images_path = "../data/images/"
masks_path = "../data/masks/"

# Grab a list of all image and mask filenames
image_files = sorted(os.listdir(images_path))
mask_files = sorted(os.listdir(masks_path))

# Let's check the first image and its corresponding mask
image_example = cv2.imread(os.path.join(images_path, image_files[0]))
mask_example = cv2.imread(os.path.join(masks_path, mask_files[0]), cv2.IMREAD_GRAYSCALE)

# Resize the image and mask to be uniform (optional, just in case)
image_example = cv2.resize(image_example, (224, 224))
mask_example = cv2.resize(mask_example, (224, 224))

# Visualize the image and mask side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(cv2.cvtColor(image_example, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
axes[0].set_title("Chest X-ray Image")
axes[1].imshow(mask_example, cmap='gray')
axes[1].set_title("Segmentation Mask")

# Show the plot
plt.show()
```
